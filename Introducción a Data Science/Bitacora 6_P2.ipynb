```python
import matplotlib.pyplot as plt
```

```python
# Top 10 países productores de vino (número de vinos producidos)
```

```python
countries_t10 = pd.DataFrame(wine_rev2.country.value_counts(dropna = True)).iloc[0:10]
countries_t10['Index'] = list(range(0,10))
countries_t10['Country_Name'] = countries_t10.index.values
countries_t10.set_index('Index',inplace = True)
countries_t10.columns = ['Number of Wines Produced','Country']
countries_t10
```

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(14,7))
sns.set(style="whitegrid")
ax = sns.barplot(y="Country", x="Number of Wines Produced", data=countries_t10)
plt.title("Top 10 Countries Producing Wine",fontsize = 25)
plt.ylabel("Country",fontsize = 20)
plt.xlabel("Number of Wines Produced",fontsize = 20)
```

### Ejercicio 3 ###

Selecciona solamente solamente los vinos realizados en Argentina, Italia y un país de su elección (recuerde usar máscaras) y realice nuevamente un scatter plot de precio vs puntaje para estos vinos. Distinga, utilizando una leyenda, la procedencia de cada vino.

```python
ar = wine_rev2[(wine_rev2.country == 'Argentina')] # Lo mismo sería: wine_rev2.loc[wine_rev2['country'] == 'Argentina']
it = wine_rev2[(wine_rev2.country == 'Italy')]
sp = wine_rev2[(wine_rev2.country == 'Spain')]

x1 = ar.points
x2 = it.points
x3 = sp.points
y1 = ar.price
y2 = it.price
y3 = sp.price

plt.plot(x1, y1, 'o', label = 'Argentina')
plt.plot(x2, y2, 'o', label = 'Italy')
plt.plot(x3, y3, 'o', label = 'Spain')

plt.title("Points and Prices in Wineries")
plt.xlabel("Points")
plt.ylabel("Prices")
plt.legend()
```
